```python
def get_guideline_display(guideline):
    if guideline == "default":
        return ""
    if guideline in CORPORA:
        guideline = CORPORA[guideline]["name"]
    return f" + {guideline}"

GUIDELINE_NAMES_IDS = {
    "default": 0,
    get_guideline_display("standard"): 1,
    get_guideline_display("general"): 2,
    get_guideline_display("rocsmt"): 3,
    get_guideline_display("footweets"): 4,
    get_guideline_display("mmtc"): 5,
    get_guideline_display("pfsmb"): 6,
}


GUIDELINES_NAMES = [
    "default",
    "standard",
    "general",
    "rocsmt",
    "footweets",
    "mmtc",
    "pfsmb"
]
```

```python
LANG_PAIRS = {
    "RoCS-MT": "En-Fr",
    "FooTweets": "En-De",
    "MMTC": "Fr-En",
    "PFSMB": "Fr-En"
}
```

```python
def get_guidelines(filename):
    guideline = filename.split(".")[-3] # guidelines.out.postproc
    if guideline not in GUIDELINES:
        guideline = "default"
    return get_guideline_display(guideline)

def get_model_name(model):
    return MODELS[model.split("/")[-1]]["name"]

def get_model_id(model):
    return MODELS[model.split("/")[-1]]["id"]

def get_guideline_id(guideline):
    if not guideline:
        return GUIDELINE_NAMES_IDS["default"]
    return GUIDELINE_NAMES_IDS[guideline]

def get_corpus_id(corpus):
    return CORPORA[corpus]["id"]

def get_corpus_name(corpus):
    if corpus == "default":
        return "None"
    return CORPORA[corpus]["name"]

def get_reference_type(corpus):
    return CORPORA[corpus]["refs"]

def get_lang_pair(corpus_name):
    return LANG_PAIRS[corpus_name]
```

```python
def normalize_error_counts(data):
    for severity in ["minor", "major", "critical", "total"]:
        data[severity] = data[severity] / data["total_sents"] * 100
    return data

def normalize_error_counts_sents(data):
    for severity in ["minor", "major", "critical", "correct"]:
        data[f"{severity}_sents"] = data[f"{severity}_sents"] / data["total_sents"] * 100
    return data
```

```python
def get_lines(corpus):
    if "rocsmt" in corpus:
        return 1922
    if "footweets" in corpus:
        return 4000
    if "mmtc" in corpus:
        return 2000
    if "pfsmb" in corpus:
        return 777
```

```python
all_data = pd.DataFrame()

for corpus in dfs:
    data = dfs[corpus].copy()
    data["guideline"] = data["file"].apply(get_guidelines)
    data["model_name"] = data["model"].apply(get_model_name) + data["guideline"] 
    data["model_id"] = data["model"].apply(get_model_id)
    data["guideline_id"] = data["guideline"].apply(get_guideline_id)
    data["corpus"] = get_corpus_name(corpus)
    data["corpus_id"] = get_corpus_id(corpus)
    # data["reference_type"] = get_reference_type(corpus)
    data["total_sents"] = get_lines(corpus)
    all_data = all_data.append(data)
```

```python
def plot_scores_by_ref(all_data, metrics, sharex=True, sharey=True):
    corpora = all_data["corpus"].unique()
    fig, axs = plt.subplots(len(corpora), len(metrics), figsize=(4*len(metrics), 3*len(corpora)), sharex=sharex, sharey=sharey)
    for i, corpus in enumerate(corpora):
        for j, metric in enumerate(metrics):
            data = all_data[all_data["corpus"] == corpus]
            sns.barplot(x=metric, y="model_name", hue="reference_type", ax=axs[i,j], data=data) #, vmin=0, vmax=100) , cmap="YlGnBu"
            axs[i,j].set_title(f"{corpus} ({get_lang_pair(corpus)})")
            axs[i,j].set_ylabel("Model (+ Guidelines)")
    plt.tight_layout()
    plt.show()
```

```python
ERROR_SEVERITY_TYPES = ["Minor", "Major", "Critical", "None"]

def plot_error_types_by_ref(all_data):
    reference_types = all_data["reference_type"].unique()
    corpora = all_data["corpus"].unique()
    fig, axs = plt.subplots(len(corpora), len(reference_types), figsize=(4*len(reference_types)+1, 3*len(corpora)), sharey=True, sharex=False)
    for i, corpus in enumerate(corpora):
        for j, ref in enumerate(reference_types):
            data = all_data[(all_data["corpus"] == corpus) & (all_data["reference_type"] == ref)]
            data = data[["model_name", "minor_sents", "major_sents", "critical_sents", "correct_sents"]]
            sns.heatmap(data.set_index("model_name"), annot=True, fmt=".0f", cmap="YlGnBu", ax=axs[i,j]) #, vmin=0, vmax=100)
            axs[i,j].set_title(f"{corpus} ({ref} references)")
            axs[i,j].set_xticklabels(ERROR_SEVERITY_TYPES, rotation=0)
            axs[i,j].set_yticklabels(data["model_name"].unique(), rotation=0)
            # reduce annotations font size
            for t in axs[i,j].texts:
                t.set_fontsize(10)
    
    # Remove individual x-axis labels
    for ax in axs.flat:
        ax.set_xlabel('')
        ax.set_ylabel('')
    
    # Remove individual y-axis labels (except for the leftmost ones)
    for ax in axs[:,1:].flat:
        ax.set_ylabel('')
    
    fig.supylabel("Model (+ Guidelines)")
    fig.supxlabel("Error Severity")
    plt.tight_layout()
    plt.show()

plot_error_types_by_ref(all_data)
```

```python
def increase_tick_fontsize(ax, size=14):
    # increase the font size of the annotations and the labels
    for t in ax.texts:
        t.set_fontsize(size)
    # increase the font size of the y-axis labels
    for t in ax.yaxis.get_ticklabels():
        t.set_fontsize(size)
    # increase the font size of the x-axis labels
    for t in ax.xaxis.get_ticklabels():
        t.set_fontsize(size)
    # increase the font size of the colorbar labels
    cbar = ax.collections[0].colorbar
    if cbar:
        cbar.ax.tick_params(labelsize=size)
    
```

```python
def plot_metric_heatmap_spaced(all_data, output_dir, metric="comet", model_names=[NLLB_MODEL_NAME, LLAMA_MODEL_NAME, GEMMA_MODEL_NAME, TOWER_MODEL_NAME], vmin=70, vmax=90):
    data = {}
    height_ratios = []
    for model in model_names:
        data[model] = all_data[all_data["model"].str.contains(model)].copy()
        data[model] = data[model].pivot(index="guideline", columns="corpus", values=metric)
        for column in data[model].columns:
            data[model].rename(columns={column: f"{column}\n({LANG_PAIRS[column]})"}, inplace=True)
        height_ratios.append(data[model].shape[0])

    # Create figure with space for colourbar
    fig, axes = plt.subplots(nrows=len(model_names), ncols=1, sharey=True, figsize=(8, sum(height_ratios)/2))
    # Create figure & GridSpec
    gs = fig.add_gridspec(len(model_names), 1, height_ratios=height_ratios)
    fig.subplots_adjust(right=0.85)  # Leave room for colourbar
    # Add a colourbar axis
    cbar_ax = fig.add_axes([1, 0.15, 0.03, 0.8])  # [left, bottom, width, height]

    # Plot each heatmap without its own colourbar
    for i, model in enumerate(model_names):
        ax = fig.add_subplot(gs[i, 0])
        sns.heatmap(
            data[model],
            ax=ax,
            annot=True,
            fmt=".2f",
            vmin=vmin,
            vmax=vmax,
            cmap="YlGnBu",
            cbar=(i == 0),       # Only the first heatmap creates a cbar
            cbar_ax=None if i != 0 else cbar_ax
        )
        ax.set_title(MODELS[model]["name"], size=14)
        increase_tick_fontsize(ax, size=16)
        ax.set_xlabel('')
        ax.set_ylabel('')
        if i < len(model_names) - 1:
            ax.set_xticklabels([])
      
    # Hide GridSpec axes
    for ax in axes:
        ax.tick_params(left=False, bottom=False, labelleft=False, labelbottom=False)
        for spine in ax.spines.values():
            spine.set_visible(False)  # remove plot frame if desired

    fig.supylabel("Guidelines", fontsize=14)
    fig.supxlabel("Corpus", fontsize=14)

    plt.tight_layout()
    plt.savefig(f"{output_dir}/{metric}_heatmap_spaced_{'_'.join(model_names)}.pdf", bbox_inches='tight')
```

```python
# plot a heatmap with the y-axis as the model and the x-axis as the corpus and the annotations as the xcomet scores
def plot_metric_heatmap(all_data, output_dir, metric="comet", model_names=[NLLB_MODEL_NAME, LLAMA_MODEL_NAME, GEMMA_MODEL_NAME, TOWER_MODEL_NAME], vmin=70, vmax=90):
    data = all_data[all_data["model"].str.contains("|".join(model_names))]
    corpora = data["corpus"].unique()
    models = data["model_name"].unique()
    fig, ax = plt.subplots(1, 1, figsize=(8,5))
    data = data.pivot(index="model_name", columns="corpus", values=metric)
    for column in data.columns:
        data.rename(columns={column: f"{column}\n({LANG_PAIRS[column]})"}, inplace=True)
    sns.heatmap(data, annot=True, fmt=".2f", cmap="YlGnBu", ax=ax, vmin=vmin, vmax=vmax)
    # ax.set_xlabel('')
    # ax.set_ylabel('')
    ax.set_ylabel("Model (+ Guidelines)", fontsize=16)
    ax.set_xlabel("Corpus", fontsize=16)
    increase_tick_fontsize(ax, size=14)
    plt.tight_layout()
    plt.savefig(f"{output_dir}/{metric}_heatmap_{'_'.join(model_names)}.pdf", format="pdf")

```

```python
ERROR_COLUMNS = ["model_name", "minor", "major", "critical", "total"]
ERROR_SENTS_COLUMNS = ["model_name", "minor_sents", "major_sents", "critical_sents", "correct_sents"]

def plot_error_types(all_data, output_dir, sentence_counts=False): 
    corpora = all_data["corpus"].unique()
    n_cols = 2
    n_rows = math.ceil(len(corpora) / n_cols)
    fig, axs = plt.subplots(n_rows, n_cols, figsize=(4*n_cols+1, 3*n_rows+1), sharey=True, sharex=False)
    for k, corpus in enumerate(corpora):
        i, j = divmod(k, n_cols)
        data = all_data[all_data["corpus"] == corpus]
        columns = ERROR_SENTS_COLUMNS if sentence_counts else ERROR_COLUMNS
        data = data[columns]
        sns.heatmap(data.set_index("model_name"), annot=True, fmt=".0f", cmap="YlGnBu", ax=axs[i,j], vmin=0, vmax=250)
        axs[i,j].set_title(f"{corpus} ({get_lang_pair(corpus)})")
        xlabels = ERROR_SEVERITY_TYPES[:3] + ["Total"] if not sentence_counts else ERROR_SEVERITY_TYPES
        axs[i,j].set_xticklabels(xlabels, rotation=0)
        # reduce annotations font size
        for t in axs[i,j].texts:
            t.set_fontsize(10)
    
    # Remove individual x-axis labels
    for ax in axs.flat:
        ax.set_xlabel('')
        ax.set_ylabel('')
    
    # Remove individual y-axis labels (except for the leftmost ones)
    for ax in axs[:,1:].flat:
        ax.set_ylabel('')
    
    fig.supylabel("Model (+ Guidelines)")
    fig.supxlabel("Detected Error Severity (xCOMET-XL)")
    plt.tight_layout()
    # save plot to output dir
    filename = "error_types_sents_heatmap.pdf" if sentence_counts else "error_types_heatmap.pdf"
    plt.savefig(f"{output_dir}/{filename}", format="pdf")
```

```python
def get_errors(errors, severity):
    sentence_ids = []
    for i, sentence in enumerate(errors):
        for span in sentence["spans"]:
            if span["severity"] == severity:
                sentence_ids.append(i)
                break
    return set(sentence_ids)

def get_correct(errors):
    sentence_ids = []
    for i, sentence in enumerate(errors):
        if len(sentence["spans"]) == 0:
            sentence_ids.append(i)
    return set(sentence_ids)
```

```python
from comet import download_model, load_from_checkpoint
from sacrebleu.metrics import BLEU
bleu_metric = BLEU(effective_order=True)
comet_model_path = download_model("Unbabel/wmt22-comet-da")
comet_model = load_from_checkpoint(comet_model_path)
```

```python
def get_bleu_scores(ref_data, sys_data):
    scores = []
    for sys, ref in zip(sys_data, ref_data):
        score = bleu_metric.sentence_score(sys, [ref]).score
        scores.append(score)
    return scores
```

```python
def get_bleu_comet_scores(src_data, ref_data, sys_data):
    bleu_result = bleu_metric.corpus_score(sys_data, [ref_data])
    data = [{"src": src, "mt": mt, "ref": ref} for src, mt, ref in zip(src_data, sys_data, ref_data)]
    comet_result = comet_model.predict(data, batch_size=32, gpus=1)
    return {
        "bleu": {
            "corpus_score": bleu_metric.corpus_score(sys_data, [ref_data]).score,
            "sentence_scores": get_bleu_scores(ref_data, sys_data)
        },
        "comet": {
            "corpus_score": comet_result.system_score,
            "sentence_scores": comet_result.scores
        }
    }
```

```python
import re
import pandas as pd
import matplotlib.pyplot as plt

GUIDELINE_ORDER = ["default", "rocsmt", "footweets", "mmtc", "pfsmb"]
GUIDELINE_LABELS = {
    "default": "None",
    "rocsmt": "+RoCS-MT",
    "footweets": "+FooTweets",
    "mmtc": "+MMTC",
    "pfsmb": "+PFSMB",
}
MODEL_LABELS = {
    "meta-llama/Llama-3.1-8B-Instruct": "Llama-3.1-8B",
    "google/gemma-2-9b-it": "Gemma-2-9B",
    "Unbabel/TowerInstruct-7B-v0.2": "Tower-7B-v0.2",
    "facebook/nllb-200-3.3B": "NLLB-3B",
}


def extract_guideline(file_name: str) -> str:
    """
    Extract guideline from filenames like:
      test.fr-en.fr.default.out.postproc
      test.fr-en.fr.rocsmt.out.postproc
      test.fr-en.fr.out.postproc   (baseline)
    """
    match = re.search(r"\.(default|footweets|mmtc|pfsmb|rocsmt)\.out\.postproc$", file_name)
    if match:
        return match.group(1)
    return "baseline"


def prepare_delta_df(csv_path: str, metric: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)

    df["guideline"] = df["file"].apply(extract_guideline)

    # Keep only non-baseline systems that actually have delta statistics
    df = df[df["guideline"].isin(GUIDELINE_ORDER)].copy()

    needed_cols = [
        "model",
        "file",
        "guideline",
        f"{metric}_delta_mean",
        f"{metric}_delta_ci_low",
        f"{metric}_delta_ci_high",
        f"{metric}_significant_95_ci",
    ]
    df = df[needed_cols].dropna(subset=[f"{metric}_delta_mean"])

    df["guideline"] = pd.Categorical(df["guideline"], GUIDELINE_ORDER, ordered=True)
    df = df.sort_values(["model", "guideline"])

    return df


def plot_delta_with_ci(csv_path: str, metric: str = "bleu", title: str | None = None):
    df = prepare_delta_df(csv_path, metric)

    fig, ax = plt.subplots(figsize=(9, 5))

    models = list(df["model"].unique())
    offsets = [-0.2, 0.0, 0.2]
    if len(models) == 4:
        offsets = [-0.27, -0.09, 0.09, 0.27]
    elif len(models) == 2:
        offsets = [-0.12, 0.12]
    elif len(models) == 1:
        offsets = [0.0]

    x_base = list(range(len(GUIDELINE_ORDER)))

    for offset, model in zip(offsets, models):
        sub = df[df["model"] == model].copy()
        sub = sub.set_index("guideline").reindex(GUIDELINE_ORDER).reset_index()

        x = [i + offset for i in x_base]
        y = sub[f"{metric}_delta_mean"].to_numpy()
        yerr_low = y - sub[f"{metric}_delta_ci_low"].to_numpy()
        yerr_high = sub[f"{metric}_delta_ci_high"].to_numpy() - y

        ax.errorbar(
            x,
            y,
            yerr=[yerr_low, yerr_high],
            fmt="o-",
            capsize=4,
            label=MODEL_LABELS.get(model, model),
        )

        # Optional: mark significant points with a star
        sig = sub[f"{metric}_significant_95_ci"].fillna(False).to_numpy()
        for xi, yi, is_sig in zip(x, y, sig):
            if is_sig:
                ax.text(xi, yi, "*", ha="center", va="bottom", fontsize=12)

    ax.axhline(0, linestyle="--", linewidth=1)

    ax.set_xticks(x_base)
    ax.set_xticklabels([GUIDELINE_LABELS[g] for g in GUIDELINE_ORDER], rotation=20)
    ax.set_ylabel(f"Δ {metric.upper()} vs NLLB")
    ax.set_xlabel("Guideline")
    ax.set_title(title or f"{metric.upper()} delta vs NLLB with 95% CI")
    ax.legend()
    fig.tight_layout()
    return fig, ax
```

```python
import os
import re
import numpy as np
import pandas as pd

GUIDELINE_ORDER = ["default", "rocsmt", "footweets", "mmtc", "pfsmb"]
GUIDELINE_LABELS = {
    "default": "None",
    "rocsmt": "+RoCS-MT",
    "footweets": "+FooTweets",
    "mmtc": "+MMTC",
    "pfsmb": "+PFSMB",
}

CORPUS_ORDER = ["rocsmt", "footweets", "mmtc", "pfsmb"]
CORPUS_LABELS = {
    "rocsmt": "RoCS-MT",
    "footweets": "FooTweets",
    "mmtc": "MMTC",
    "pfsmb": "PFSMB",
}

MODEL_ORDER = [
    "facebook/nllb-200-3.3B",
    "meta-llama/Llama-3.1-8B-Instruct",
    "google/gemma-2-9b-it",
    "Unbabel/TowerInstruct-7B-v0.2",
]
MODEL_LABELS = {
    "facebook/nllb-200-3.3B": "NLLB-3B",
    "meta-llama/Llama-3.1-8B-Instruct": "Llama-3.1-8B",
    "google/gemma-2-9b-it": "Gemma-2-9B",
    "Unbabel/TowerInstruct-7B-v0.2": "Tower-7B-v0.2",
}


def extract_guideline(file_name: str) -> str:
    match = re.search(r"\.(default|footweets|mmtc|pfsmb|rocsmt)\.out\.postproc$", file_name)
    if match:
        return match.group(1)
    return "baseline"


def significance_marker(is_sig):
    return "*" if bool(is_sig) else ""


def arrow_latex(delta):
    if pd.isna(delta):
        return ""
    if delta > 0:
        return r"\colorbox{mygreen}{$\uparrow$}"
    if delta < 0:
        return r"\colorbox{myred}{$\downarrow$}"
    return ""


def load_all_metric_rows(score_dir: str) -> pd.DataFrame:
    frames = []
    for corpus in CORPUS_ORDER:
        csv_path = os.path.join(score_dir, f"scores_ci_{corpus}.csv")
        df = pd.read_csv(csv_path)
        df["guideline"] = df["file"].apply(extract_guideline)
        df["corpus"] = corpus
        frames.append(df)
    return pd.concat(frames, ignore_index=True)


def build_metric_table_compact(score_dir: str, metric: str = "bleu"):
    df = load_all_metric_rows(score_dir)

    score_col = metric
    delta_col = f"{metric}_delta_mean"
    sig_col = f"{metric}_significant_95_ci"

    rows = []

    for model in MODEL_ORDER:
        model_df = df[df["model"] == model].copy()
        model_label = MODEL_LABELS.get(model, model)

        if model == "facebook/nllb-200-3.3B":
            row = {"Model": model_label, "Guideline": "—"}

            for corpus in CORPUS_ORDER:
                corpus_df = model_df[
                    (model_df["corpus"] == corpus) & (model_df["guideline"] == "baseline")
                ]
                value = corpus_df.iloc[0][score_col] if not corpus_df.empty else np.nan
                row[CORPUS_LABELS[corpus]] = f"{value:.2f}" if pd.notna(value) else ""

            rows.append(row)
            continue

        for i, guideline in enumerate(GUIDELINE_ORDER):
            row = {
                "Model": model_label if i == 0 else "",
                "Guideline": GUIDELINE_LABELS[guideline],
            }

            for corpus in CORPUS_ORDER:
                sub = model_df[
                    (model_df["corpus"] == corpus) & (model_df["guideline"] == guideline)
                ]

                if sub.empty or pd.isna(sub.iloc[0].get(score_col, np.nan)):
                    row[CORPUS_LABELS[corpus]] = ""
                else:
                    r = sub.iloc[0]
                    arrow = arrow_latex(r.get(delta_col, np.nan))
                    marker = significance_marker(r.get(sig_col, False))
                    row[CORPUS_LABELS[corpus]] = f"{r[score_col]:.2f}{arrow}{marker}"

            rows.append(row)

    table_df = pd.DataFrame(rows)

    ordered_cols = ["Model", "Guideline"] + [CORPUS_LABELS[c] for c in CORPUS_ORDER]
    table_df = table_df[ordered_cols]

    return table_df

```

```python
# Example usage
metric = "cometkiwi"

table_df = build_metric_table_compact(score_dir, metric=metric)

table_df.to_latex(
    f"{score_dir}/{metric}_compact_table.tex",
    index=False,
    escape=False,
    column_format="ll" + "c" * 4,
)
```

```python
import os
import re
import pandas as pd
import matplotlib.pyplot as plt

# Camera-ready defaults
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 11,
    "axes.titlesize": 12,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 11,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})

GUIDELINE_ORDER = ["rocsmt", "footweets", "mmtc", "pfsmb"]
GUIDELINE_LABELS = {
    "rocsmt": "+RoCS-MT",
    "footweets": "+FooTweets",
    "mmtc": "+MMTC",
    "pfsmb": "+PFSMB",
}

CORPUS_ORDER = ["rocsmt", "footweets", "mmtc", "pfsmb"]
CORPUS_LABELS = {
    "rocsmt": "RoCS-MT",
    "footweets": "FooTweets",
    "mmtc": "MMTC",
    "pfsmb": "PFSMB",
}

MODEL_LABELS = {
    "meta-llama/Llama-3.1-8B-Instruct": "Llama-3.1-8B",
    "google/gemma-2-9b-it": "Gemma-2-9B",
    "Unbabel/TowerInstruct-7B-v0.2": "Tower-7B-v0.2",
    "facebook/nllb-200-3.3B": "NLLB-3B",
}

# Camera-ready matplotlib defaults
plt.rcParams.update({
    "font.family": "serif",
    "font.size": 11,
    "axes.titlesize": 12,
    "axes.labelsize": 11,
    "xtick.labelsize": 10,
    "ytick.labelsize": 10,
    "legend.fontsize": 12,
    "pdf.fonttype": 42,
    "ps.fonttype": 42,
})
def extract_guideline(file_name: str) -> str:
    match = re.search(r"\.(default|footweets|mmtc|pfsmb|rocsmt)\.out\.postproc$", file_name)
    if match:
        return match.group(1)
    return "baseline"


def prepare_delta_vs_default_df(csv_path: str, metric: str, model: str) -> pd.DataFrame:
    df = pd.read_csv(csv_path)
    df["guideline"] = df["file"].apply(extract_guideline)

    needed_cols = [
        "model",
        "file",
        "guideline",
        f"{metric}_delta_mean",
        f"{metric}_delta_ci_low",
        f"{metric}_delta_ci_high",
        f"{metric}_significant_95_ci",
    ]

    df = df[df["model"] == model].copy()
    df = df[df["guideline"].isin(GUIDELINE_ORDER)].copy()
    df = df[needed_cols].dropna(subset=[f"{metric}_delta_mean"])

    df["guideline"] = pd.Categorical(df["guideline"], GUIDELINE_ORDER, ordered=True)
    df = df.sort_values("guideline")
    return df


def plot_delta_vs_default_1x4(
    model: str,
    metric: str,
    score_dir: str,
    output_path: str | None = None,
):
    fig, axes = plt.subplots(1, 4, figsize=(10.5, 3.8), sharey=True)

    for ax, corpus in zip(axes, CORPUS_ORDER):
        csv_path = os.path.join(score_dir, f"scores_ci_default_{corpus}.csv")
        df = prepare_delta_vs_default_df(csv_path, metric, model)

        df = df.set_index("guideline").reindex(GUIDELINE_ORDER).reset_index()

        x = list(range(len(GUIDELINE_ORDER)))
        y = df[f"{metric}_delta_mean"].to_numpy()
        yerr_low = y - df[f"{metric}_delta_ci_low"].to_numpy()
        yerr_high = df[f"{metric}_delta_ci_high"].to_numpy() - y

        ax.errorbar(
            x,
            y,
            yerr=[yerr_low, yerr_high],
            fmt="o-",
            linewidth=1.5,
            markersize=5,
            capsize=3,
        )

        sig = df[f"{metric}_significant_95_ci"].fillna(False).to_numpy()
        for xi, yi, is_sig in zip(x, y, sig):
            if is_sig:
                ax.text(xi, yi, "*", ha="center", va="bottom", fontsize=11)

        ax.axhline(0, linestyle="--", linewidth=1)
        ax.set_xticks(x)
        ax.set_xticklabels([GUIDELINE_LABELS[g] for g in GUIDELINE_ORDER], rotation=20)
        ax.set_title(CORPUS_LABELS[corpus], pad=6)
        ax.set_xlabel("Guideline")

    fig.supylabel(f"Δ {metric.upper()} vs default", fontsize=12, x=0.04)
    # fig.suptitle(f"{MODEL_LABELS.get(model, model)}", fontsize=13, y=1.03)

    fig.tight_layout(rect=[0.04, 0.02, 1.0, 0.95], pad=0.6, w_pad=1.0)

    if output_path is not None:
        fig.savefig(output_path, format="pdf", bbox_inches="tight", pad_inches=0.02)

    return fig, axes
```

```python
import os
import re
import numpy as np
import pandas as pd

GUIDELINE_ORDER = ["default", "rocsmt", "footweets", "mmtc", "pfsmb"]
GUIDELINE_LABELS = {
    "default": "None",
    "rocsmt": "+RoCS-MT",
    "footweets": "+FooTweets",
    "mmtc": "+MMTC",
    "pfsmb": "+PFSMB",
}

CORPUS_ORDER = ["rocsmt", "footweets", "mmtc", "pfsmb"]
CORPUS_LABELS = {
    "rocsmt": "RoCS-MT",
    "footweets": "FooTweets",
    "mmtc": "MMTC",
    "pfsmb": "PFSMB",
}

MODEL_ORDER = [
    "meta-llama/Llama-3.1-8B-Instruct",
    "google/gemma-2-9b-it",
    "Unbabel/TowerInstruct-7B-v0.2",
]
MODEL_LABELS = {
    "meta-llama/Llama-3.1-8B-Instruct": "Llama-3.1-8B",
    "google/gemma-2-9b-it": "Gemma-2-9B",
    "Unbabel/TowerInstruct-7B-v0.2": "Tower-7B-v0.2",
}


def extract_guideline(file_name: str) -> str:
    match = re.search(r"\.(default|footweets|mmtc|pfsmb|rocsmt)\.out\.postproc$", file_name)
    if match:
        return match.group(1)
    return "baseline"


def significance_marker(is_sig):
    return "*" if bool(is_sig) else ""


def arrow_latex(delta):
    if pd.isna(delta):
        return ""
    if delta > 0:
        return r"\colorbox{mygreen}{$\uparrow$}"
    if delta < 0:
        return r"\colorbox{myred}{$\downarrow$}"
    return ""


def load_all_metric_rows(score_dir: str) -> pd.DataFrame:
    frames = []
    for corpus in CORPUS_ORDER:
        csv_path = os.path.join(score_dir, f"scores_ci_default_{corpus}.csv")
        df = pd.read_csv(csv_path)
        df["guideline"] = df["file"].apply(extract_guideline)
        df["corpus"] = corpus
        frames.append(df)
    return pd.concat(frames, ignore_index=True)


def build_metric_table_compact_vs_default(score_dir: str, metric: str = "bleu"):
    df = load_all_metric_rows(score_dir)

    score_col = metric
    delta_col = f"{metric}_delta_mean"
    sig_col = f"{metric}_significant_95_ci"

    rows = []

    for model in MODEL_ORDER:
        model_df = df[df["model"] == model].copy()
        model_label = MODEL_LABELS.get(model, model)

        for i, guideline in enumerate(GUIDELINE_ORDER):
            row = {
                "Model": model_label if i == 0 else "",
                "Guideline": GUIDELINE_LABELS[guideline],
            }

            for corpus in CORPUS_ORDER:
                sub = model_df[
                    (model_df["corpus"] == corpus) & (model_df["guideline"] == guideline)
                ]

                if sub.empty or pd.isna(sub.iloc[0].get(score_col, np.nan)):
                    row[CORPUS_LABELS[corpus]] = ""
                    continue

                r = sub.iloc[0]

                # Default is the within-model reference: show score only, no marker.
                if guideline == "default":
                    row[CORPUS_LABELS[corpus]] = f"{r[score_col]:.2f}"
                else:
                    arrow = arrow_latex(r.get(delta_col, np.nan))
                    marker = significance_marker(r.get(sig_col, False))
                    row[CORPUS_LABELS[corpus]] = f"{r[score_col]:.2f}{arrow}{marker}"

            rows.append(row)

    table_df = pd.DataFrame(rows)
    ordered_cols = ["Model", "Guideline"] + [CORPUS_LABELS[c] for c in CORPUS_ORDER]
    table_df = table_df[ordered_cols]
    return table_df

```
